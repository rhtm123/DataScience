```python
# find the top5 bastmans in the world from 2009 to 2019
# based on average, strike rate (atleast 20 matches are played by a batsman)

import pandas as pd
```

```python
df = pd.read_csv('../sample_data/Batsman_Data.csv')
```

```python
df.sample(5)
```

```python
df.dtypes
```

```python
# Remove the DNB
# d = df.set_index('Bat1')
# # d
# df_new = d.drop(["DNB","TDNB"], axis=0)

# df_new.reset_index()
# df = df_new

### filter

f1 = df['Bat1'] !='DNB'
f2 = df['Bat1'] !='TDNB'
f3 = df['Runs'] != '-'

df_n= df[f1 & f2 & f3]
df_n
# df.sample(10)
# f3
```

```python
## changing the data types
df_n = df_n.astype({'Runs':'int','4s':'int','6s':'int','BF':'int'})
```

```python
## Group By 
d_count = df_n.groupby("Batsman").count()
d_sum = df_n.groupby("Batsman").sum()
```

```python
d_sum['NumMatch'] = d_count['Bat1']
d_sum['Avg'] = d_sum['Runs']/d_sum['NumMatch']
d_sum['StrikeRate'] = d_sum['Runs']/d_sum['BF']*100
d_final = d_sum[d_sum['NumMatch']>20]
```

```python
d_final
```

```python
df_sorted_avg = d_final.sort_values('Avg', ascending=False)
df_sorted_sr = d_final.sort_values('StrikeRate', ascending=False)
```
