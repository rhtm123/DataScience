<h2 style="color:blue"> WordCloud </h2>

Word Cloud is a data visualization technique used for representing text <br>
data in which the size of each word indicates its frequency or importance. <br>
Significant textual data points can be highlighted using a word cloud. 


```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
file = open('sample_data/imran_khan_un_speech_2019')
```

```python
lines = [line.strip() for line in file.readlines()]

# print(lines)

speech = " ".join(lines)
```

```python
speech
```

```python
# print(STOPWORDS)

wc_imran = WordCloud(width = 800, height = 800, background_color ='white', stopwords = STOPWORDS)
```

```python
wc_imran.generate(speech)
```

```python
file_modi = open('sample_data/modi_speech_un_2019')
```

```python
lines = [line.strip() for line in file_modi.readlines()]

speech_modi = " ".join(lines)
```

```python
# speech_modi
```

```python
wc_modi = WordCloud(width = 800, height = 800, background_color ='white', stopwords = STOPWORDS)
```

```python
wc_modi.generate(speech_modi)
```

```python
plt.figure(figsize = (8, 8), facecolor = None) 
plt.title("Imran Khan Speech", fontsize=30, pad=30)
plt.imshow(wc_imran) 
plt.axis("off") 
plt.tight_layout(pad = 0)
# plt.savefig('imran_khan_speech_un_2019_word_cloud.pdf')
plt.show()
```
