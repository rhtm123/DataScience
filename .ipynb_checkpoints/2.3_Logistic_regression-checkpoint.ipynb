```python
## Data Cleaning

def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age

train['Age'] = train[['Age','Pclass']].apply(impute_age, axis=1)
```

```python
# train.drop(['Cabin'], axis=1, inplace=True)

train.dropna(inplace=True)
train.count()
```
